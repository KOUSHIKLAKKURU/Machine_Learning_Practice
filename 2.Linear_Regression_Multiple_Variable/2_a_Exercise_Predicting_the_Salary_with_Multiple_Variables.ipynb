```python
import numpy as mp
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from word2number import w2n
```

```python
df=pd.read_csv("hiring.csv")
df
```

**Preprocessing the dataset**

Converting experience values to integers and filling the Nan values to zero experience

```python
df.experience=df.experience.fillna("zero")
df
```

```python
df.experience=df.experience.apply(w2n.word_to_num)
df
```

Replacing test_score NaN values to median

```python
import math
median_test=math.floor(df['test_score(out of 10)'].mean())
median_test
```

```python
df['test_score(out of 10)']=df['test_score(out of 10)'].fillna(median_test)
df
```

**Training the model**

```python
reg=linear_model.LinearRegression()
reg.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']],df[['salary($)']])
```

```python
reg.predict([[2,9,6]])
```
