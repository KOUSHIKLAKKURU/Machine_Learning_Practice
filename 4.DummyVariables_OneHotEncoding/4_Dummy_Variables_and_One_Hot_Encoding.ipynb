```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("homeprices.csv")
df
```

**Creating Dummy Variables or Columns using pandas get_dummies method**

```python
dummies=pd.get_dummies(df.town)
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
merged
```

Dropping town column and one dummy variable column from merged dataframe

```python
final=merged.drop(['town','west windsor'],axis='columns')
final
```

```python
model=linear_model.LinearRegression()
```

```python
X=final.drop('price',axis='columns')
X
```

```python
y=final.price
y
```

```python
model.fit(X,y)
```

**Predicting the price for 2800 area in robinsville town**

```python
model.predict([[2800,0,1]])
```

**Predicting the price for 3400 area in west windsor town**

```python
model.predict([[3400,0,0]])
```

```python
model.score(X,y)
```

**Creating Dummy Variables or Columns using sklearn one hot encoding method**

```python
df
```

In order to do one hot encoding we need to perform label encoding to town column. Label Encoding refers to assigning the string variables to some integer values.

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
dfle=df
dfle.town=le.fit_transform(dfle.town) # fit_transform() method takes the column and label encodes it.
dfle
```

```python
X=dfle[['town','area']].values
X
```

```python
y=dfle.price
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([("town", OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
X=X[:,1:] # This will drop one dummy variable column
X
```

```python
model.fit(X,y)
```

```python
model.predict([[1,0,2800]])
```

```python
model.predict([[0,1,3400]])
```
